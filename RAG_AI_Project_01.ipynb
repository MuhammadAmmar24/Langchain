```python
from langchain_community.document_loaders import GithubFileLoader
```

```python
# Loading Github Repo
loader = GithubFileLoader(
    repo="panaversity/learn-applied-generative-ai-fundamentals",  # the repo name
    branch="main",  # the branch name
    access_token="YOU_ACCESS_TOKEN",
    github_api_url="https://api.github.com",
    file_filter=lambda file_path: file_path.endswith(
        ".md"
    ),  # load all markdowns files.
)

docs = loader.load()

```

```python
print(len(docs))
print(docs[73].metadata)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

#Splitting

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)

splits = text_splitter.split_documents(docs)
```

```python
print(len(splits))
```

```python
from langchain_huggingface.embeddings import HuggingFaceEmbeddings
from langchain_chroma import Chroma
```

```python
embeddings = HuggingFaceEmbeddings()
```

```python
vectorstore = Chroma.from_documents(documents=splits, embedding=embeddings)
```

```python
print(vectorstore)
```

```python
# Retrieve and generate using the relevant snippets of the blog.
retriever = vectorstore.as_retriever(
    search_type="similarity_score_threshold",  search_kwargs={"score_threshold": 0.2}
)

docs = retriever.invoke("what is regression model?")
print(len(docs))
print(docs)

```

```python
# Initialize LLM

from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(model="gemini-1.5-flash", google_api_key="GOOGLE_KEY")
print(llm)
```

```python
from langchain import hub
```

```python
prompt = hub.pull("rlm/rag-prompt")
```

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough


def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)


rag_chain = (
    {"context": retriever | format_docs, "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)

rag_chain.invoke("how to build AI Agents?")
```
